## Problema

En entornos organizacionales, los objetivos estratégicos suelen formularse de manera abstracta, por ejemplo:

- “Mejorar la eficiencia operativa”
- “Reducir tiempos de preparación”
- “Optimizar el servicio al cliente”

Sin embargo, estos objetivos no siempre se traducen automáticamente en indicadores cuantificables y accionables.

Esto genera:

- Dificultad para medir resultados
- Interpretaciones ambiguas
- Falta de criterios objetivos de seguimiento
- Dependencia del análisis manual

Aunque los modelos de Inteligencia Artificial pueden ayudar en el análisis, cuando se utilizan con prompts poco estructurados tienden a generar respuestas generales y poco aplicables.

Por lo tanto, el problema no es la falta de tecnología, sino la falta de diseño estratégico en la construcción del prompt.

Este proyecto busca resolver ese problema.

## Propuesta de Solución

Se propone desarrollar una Proof of Concept (POC) que permita transformar objetivos estratégicos abstractos en indicadores cuantificables y estructurados utilizando técnicas de Fast Prompting.

La solución consiste en diseñar dos tipos de prompts:

1. Un Prompt Básico, sin estructura definida.
2. Un Prompt Optimizado (Fast Prompt), que incorpora:
   - Asignación de rol
   - Definición explícita del objetivo
   - Estructura obligatoria de salida
   - Identificación de variable cuantificable
   - Fórmula matemática explícita
   - Clasificación estratégica del indicador

A través de la comparación entre ambos enfoques, se demuestra cómo el diseño estructurado del prompt impacta directamente en la calidad del resultado generado.

Además, se incorpora un modelo de generación de imagen para representar visualmente el flujo:

Objetivo → Datos → Análisis → KPI → Decisión

De esta manera, el proyecto integra modelos texto-texto y texto-imagen para resolver la problemática identificada.

## Implementación

La Proof of Concept (POC) fue desarrollada en Python dentro de un Jupyter Notebook.

Se implementaron dos enfoques para resolver el problema:

1. Prompt Básico  
2. Prompt Optimizado (Fast Prompt)

El objetivo de la implementación es demostrar cómo el diseño estructurado del prompt impacta directamente en la calidad del resultado generado.

### Estructura del Código

El sistema incluye:

- Función `prompt_basico()`  
  Genera una respuesta general sin estructura formal.

- Función `prompt_fast()`  
  Genera una salida estructurada que incluye:
  - Objetivo analizado
  - Problema cuantificable
  - Indicador propuesto
  - Fórmula matemática
  - Dimensión estratégica

- Función de evaluación estructural  
  Permite comparar el nivel de organización del resultado.

- Simulación de conteo de tokens  
  Permite analizar la optimización del prompt.

```python
def prompt_basico(objetivo):
    return f"""
    Indicador sugerido para {objetivo}:
    Medir desempeño general del proceso.
    """
```

```python
def prompt_fast(objetivo):

    return {
        "objetivo_analizado": objetivo,
        "problema_detectado": "Falta definición de variable cuantificable",
        "indicador_propuesto": "Tiempo promedio por pedido",
        "formula": "Tiempo total preparación / cantidad de pedidos",
        "dimension": "Eficiencia"
    }
```

```python
objetivo = "Reducir tiempos de preparación en depósito"

print("=== Prompt Básico ===")
print(prompt_basico(objetivo))

print("\n=== Prompt Fast ===")
print(prompt_fast(objetivo))
```

## Análisis Comparativo

El Prompt Básico devuelve una respuesta general y poco estructurada.  
No controla formato ni obliga al modelo a seguir pasos definidos.

El Prompt Fast aplica técnicas de Fast Prompting:

- Asignación de rol
- Secuencia de pasos obligatorios
- Estructura de salida en formato JSON
- Claridad en variables y dimensión del indicador

Esto permite:

✔ Mayor consistencia  
✔ Mejor control del resultado  
✔ Reducción de ambigüedad  
✔ Menor necesidad de iteraciones  

Conclusión: La estructuración del prompt mejora la calidad y reutilización del output.

### Prompt básico

```python
def contar_tokens_simulado(texto):
    return len(texto.split())

basico_output = prompt_basico(objetivo)
fast_output = str(prompt_fast(objetivo))

print("Tokens Prompt Básico:", contar_tokens_simulado(basico_output))
print("Tokens Prompt Fast:", contar_tokens_simulado(fast_output))
```

### Fast Prompt

#### Técnicas de Fast Prompting Aplicadas

- Role Prompting
- Structured Output (JSON)
- Secuenciación obligatoria de pasos
- Claridad en variables cuantificables
- Control de formato de salida

Estas técnicas permiten reducir ambigüedad y mejorar la reutilización del resultado.

```python
def evaluar_estructura(respuesta):
    if isinstance(respuesta, dict):
        return "Alta estructura"
    else:
        return "Baja estructura"

print("Estructura Básico:", evaluar_estructura(prompt_basico(objetivo)))
print("Estructura Fast:", evaluar_estructura(prompt_fast(objetivo)))
```

### Comparativa

```python
def iteracion_mejora(objetivo):
    resultado_1 = prompt_basico(objetivo)
    resultado_2 = prompt_fast(objetivo)
    
    return {
        "Iteración 1": resultado_1,
        "Iteración 2 (Optimizada)": resultado_2
    }

print(iteracion_mejora(objetivo))
```

## Evolución Iterativa del Prompt

La Iteración 1 genera una respuesta conceptual sin estructura técnica.

La Iteración 2 incorpora técnicas de Fast Prompting, produciendo:

- Identificación del problema
- Definición de KPI cuantificable
- Fórmula matemática explícita
- Clasificación estratégica (dimensión)
- Formato estructurado reutilizable

Esto demuestra que el refinamiento del prompt mejora la calidad, precisión y aplicabilidad del resultado.

## Conclusión Técnica

El Prompt Básico genera respuestas genéricas con bajo nivel de estructuración, lo que dificulta su reutilización en procesos automatizados.

El Prompt Fast, aunque utiliza mayor cantidad de tokens, produce una salida estructurada y cuantificable. Esto permite:

- Integración directa en sistemas
- Automatización de análisis
- Reducción de ambigüedad

La técnica de Fast Prompting mejora la calidad operativa del resultado, priorizando estructura y precisión por sobre brevedad.
